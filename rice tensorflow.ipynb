# **Import Dataset**

```python
Train = keras.utils.image_dataset_from_directory(
    directory = 'C:/Users/M.Rezajoo/Desktop/rice/train',
    labels = 'inferred',
    label_mode = 'categorical',
    batch_size = 32,
    image_size = (250,250))    # Images resolution
```

```python
Test = keras.utils.image_dataset_from_directory(
    directory = 'C:/Users/M.Rezajoo/Desktop/rice/test',
    labels = 'inferred',
    label_mode = 'categorical',
    batch_size = 32,
    image_size = (250,250))
```

```python
Validation = keras.utils.image_dataset_from_directory(
    directory = 'C:/Users/M.Rezajoo/Desktop/rice/validation',
    labels = 'inferred',
    label_mode = 'categorical',
    batch_size = 32,
    image_size = (250,250))
```

# **CNN Model 1**

```python
CNN1 = tf.keras.models.Sequential()
```

```python
CNN1.add(tf.keras.Input(shape=[250, 250, 3]))

CNN1.add(Conv2D(32, (3,3) , kernel_initializer = 'random_normal'))
CNN1.add(Activation('relu'))

CNN1.add(Conv2D(32, (3,3) , kernel_initializer = 'random_normal'))
CNN1.add(Activation('relu'))

CNN1.add(GlobalMaxPooling2D())
CNN1.add(Dense(500))     #fully connected
CNN1.add(Dropout(0.25))
CNN1.add(Dense(5))  #output num_classes = 5
CNN1.add(Activation('softmax'))
CNN1.summary()
```

```python
CNN1.compile(optimizer = 'adam' , loss = 'categorical_crossentropy',metrics = ['accuracy'])
```

```python
model1 = CNN1.fit(x = Train , validation_data = Validation , epochs = 10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize = (10,6))
sns.set_style('whitegrid')
plt.plot(model1.history['accuracy'] , color = '#E74C3C' , marker = 'o')
plt.plot(model1.history['val_accuracy'] , color = '#641E16' , marker = 'h')
plt.title('Accuracy comparison between Validation and Train data' , fontsize = 15)
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train' , 'Test'] , loc = 'lower right')
plt.show()
```

```python
plt.figure(figsize = (10,6))
sns.set_style('whitegrid')
plt.plot(model1.history['loss'] , color = 'Purple' , marker = 'o')
plt.plot(model1.history['val_loss'] , color = 'Orange' , marker = 'h')
plt.title('Loss comparison between Validation and Train data' , fontsize = 15)
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train' , 'Test'] , loc = 'lower left')
plt.show()
```

# **CNN Model 2**

```python
CNN2 = tf.keras.models.Sequential()
```

```python
CNN2.add(tf.keras.Input(shape=[250, 250, 3]))

CNN2.add(Conv2D(32, (3,3) , kernel_initializer = 'random_uniform'))
CNN1.add(Activation('relu'))

CNN2.add(Conv2D(32, (3,3) , kernel_initializer = 'random_uniform'))
CNN1.add(Activation('relu'))

CNN2.add(Conv2D(48, (3,3) , kernel_initializer = 'random_uniform'))
CNN2.add(Activation('relu'))
CNN2.add(BatchNormalization())

CNN2.add(GlobalMaxPooling2D())
CNN2.add(Dense(500))     #fully connected
CNN2.add(Dropout(0.25))
CNN2.add(Dense(5))  #output num_classes = 5
CNN2.add(Activation('softmax'))
CNN2.summary()
```

```python
CNN2.compile(optimizer = 'RMSProp' , loss = 'categorical_crossentropy',metrics = ['accuracy'])
```

```python
model2 = CNN2.fit(x = Train , validation_data = Validation , epochs = 10)
```
